```python
# 16. Write a Python program to find the GCD (Greatest Common Divisor) of two numbers using a for loop.

def factor(num):
    f_list = []
    for i in range(1,num+1):
        if num % i == 0 :
            f_list.append(i)
    return f_list

def gcd(num1, num2):
    gcd_list = []
    for i in factor(num1):
        if i in factor(num2):
            gcd_list.append(i)
    return sorted(gcd_list)[-1]

num1, num2 = 24, 36
print(f"the greatest common divisor (GCD) of {num1} and {num2} is: {gcd(num1, num2)}")
```

```python
# 23. Write a Python program to find the LCM (Least Common Multiple) of two numbers using a for loop.

def factor(num):
    f_list = []
    for i in range(1,num+1):
        if num % i == 0 :
            f_list.append(i)
    return f_list

def gcd(num1, num2):
    gcd_list = []
    for i in factor(num1):
        if i in factor(num2):
            gcd_list.append(i)
    return sorted(gcd_list)[-1]

def LCM(num1, num2):
    if num1 == 0 or num2 == 0:
        return 0

    gcd_var = gcd(num1, num2)
    lcm = abs(num1 * num2) // gcd_var
    return lcm

num1, num2 = 24, 36
print(f"the LCM (Least Common Multiple) of {num1} and {num2} is: {LCM(num1, num2)}")
```

# **Map**

**1. Explain the purpose of the `map()` function in Python and provide an example of how it can be used to apply a function to each element of an iterable.**

The map() function in Python is a built-in function that is used to apply a specified function to all the items in an iterable (such as a list) and return an iterator that produces the results. The map() function takes a function and one or more iterables as arguments. It then applies the function to each corresponding element of the iterables and returns an iterator of the results. If multiple iterables are provided, the function should take as many arguments as there are iterables. provides a concise way to perform the same operation on each element of an iterable without the need for explicit loops.

**e.g.,**

```python
map_list = [1,2,3,4,5,6]

sq_num = lambda element: element ** 2

sqrd = list(map(sq_num, map_list))
print(f"after squaring the numbers: {sqrd}")
```

**3. How does the `map()` function differ from a list comprehension in Python, and when would you choose one over the other?**

1. * map() is a higher-order function that takes a function and one or more iterables as arguments. It returns an iterator.
            Syntax: result_map = map(function, iterable)
   * List comprehension is a concise way to create lists by applying an expression to each element of an iterable.
            Syntax: result_list_comprehension = [expression for item in iterable]
            
2. * map() returns an iterator, so you need to convert it to a list or another iterable type if you want to see the results immediately.
   * List comprehension returns a new list directly
   
3. * map() can be useful for more complex operations or when you have multiple iterables, but it may be less readable than list comprehensions for simple operations.
   * List comprehensions are often considered more readable and Pythonic, especially for straightforward operations.
   
4. * In some cases, map() may offer a slight performance advantage, especially when dealing with built-in functions, as it is implemented in C.
   * List comprehensions are generally very efficient, and any performance differences may be negligible for most use cases.
   
   
**> Choosing One Over the Other:**
   * Use map() when you have a pre-existing function (or a built-in function) and want to apply it to each element of one or more iterables.
   * Use list comprehensions when you want to create a new list by applying an expression to each element of an iterable, especially for simple operations where readability is a priority.

**1. What is the `reduce()` function in Python, and what module should you import to use it? Provide an example of its basic usage.**

The reduce() function in Python is part of the functools module. It is used to cumulatively apply a binary function (a function that takes two arguments) to the items of an iterable, from left to right, so as to reduce the iterable to a single accumulated result. The reduce() function is particularly useful when you want to perform operations such as summing elements, finding the maximum/minimum value, or accumulating results based on a custom binary function.

**Syntax: functools.reduce(function, iterable[, initializer]).** 

**e.g.,**

```python
import functools

list_red = [1,2,3,4,5]
add_sum = lambda x, y: x+y

result = functools.reduce(add_sum, list_red)
print(f"After adding the element: {result}")
```

# **Recursion:-**

**1. Explain the concept of recursion in Python. How does it differ from iteration?**


Recursion is a programming concept where a function calls itself directly or indirectly to solve a problem. In a recursive function, the problem is divided into smaller subproblems, and the function calls itself to solve these subproblems. Each recursive call works on a smaller instance of the original problem until a base case is reached, at which point the recursion stops.

* Differences between recursion and iteration:
1. Recursion involves a function calling itself But Iteration involves a loop construct (e.g., for or while) to repeat a block of code.
2. Recursion relies on the call stack and function calls.Otherside, Iteration relies on loop control structures.
3. Recursion can be elegant and easy to understand for certain problems. And same time, Iteration is often more explicit and familiar for some programmers.
4. Recursion may use more memory due to the call stack, which can lead to a stack overflow for very deep recursion.But Iteration typically uses less memory, as it doesn't involve stacking function calls.
5. In some cases, recursion may have slightly more overhead due to function calls. But Iteration may be more efficient in terms of performance for certain tasks.

The base case is a condition that, when met, stops the recursive calls and allows the function to return a result. If you omit the base case or make a mistake in defining it, the recursive function may run indefinitely, leading to a stack overflow error. It's crucial to carefully design the termination conditions to ensure that the recursion stops at some point.

```python
# 9. Write a recursive Python function to find all permutations of a given string.

def find_permutations(s):
    if len(s) <= 1:
        return [s]

    rest_permutations = find_permutations(s[1:])

    all_permutations = []
    for perm in rest_permutations:
        for i in range(len(perm) + 1):
            all_permutations.append(perm[:i] + s[0] + perm[i:])

    return all_permutations

input_string = "sou"
result = find_permutations(input_string)
print(f"Permutations of '{input_string}': {result}")
```

```python
# 10. Write a recursive Python function to check if a string is a palindrome.

def str_palindrome(string):
    if len(string) <= 1:
        return True
    return string[0] == string[-1] and str_palindrome(string[1:-1]) #excluding the first and last characters by writing like: [1:-1]

string = 'eye'
print(f"is a given string '{string}' palindrome: {str_palindrome(string)}")
```

```python
# 11. Create a recursive Python function to generate all possible combinations of a list of elements.

def get_combinations(elements):
    if not elements:
        return [[]]

    rest_combinations = get_combinations(elements[1:])

    current_element = elements[0]
    combinations_with_current = []

    for combination in rest_combinations:
        combinations_with_current.append([current_element] + combination)
        combinations_with_current.append(combination)

    return combinations_with_current

input_elements = [1, 2, 3]
result = get_combinations(input_elements)
print(f"all possible combinations of a list of elements: {result}")
```

**2. Describe some best practices for naming functions and variables in python including naming conventions and guidelines**

Naming functions and variables in Python follows certain conventions and guidelines to enhance code readability and maintainability.

**Variables:**
* Use descriptive and meaningful names that reflect the purpose or content of the variable.
* Use snake_case for variable names (lowercase with underscores between words). Example: user_name, total_count, average_value.
* Avoid using single-letter variable names unless they represent common conventions (e.g., i for an index in a loop).
* Avoid using Python reserved words as variable names.
* Be consistent with your naming style throughout the codebase.

**Functions:**
* Use clear and descriptive names for functions that indicate their purpose or action.
* Follow the snake_case convention for function names. Example: calculate_average, fetch_data_from_api.
* Choose function names that form a verb-noun pair, describing the action the function performs. Example: calculate_average, validate_input.
* Special methods (dunder methods) should have double leading and trailing underscores but avoid using underscores elsewhere in function names. Example: __init__, __str__.
* Maintain consistency in function naming across your codebase.

**General Guidelines:**
* Avoid ambiguous or unclear abbreviations that might confuse others. Example: Instead of num_items, use number_of_items.
* Prioritize meaningful names over brevity. Example: calculate_average is better than calc_avg.
* Avoid using Hungarian Notation (e.g., prefixing variable names with the type information) in Python.
* Avoid using magic numbers (hard-coded numeric values without explanation). Use named constants instead.
* Keep functions concise and focused on a single responsibility.
* Provide docstrings for public functions to explain their purpose, parameters, and expected return values.
* If a variable represents a collection, be consistent with using singular or plural names. Example: user vs. users, item vs. items.
